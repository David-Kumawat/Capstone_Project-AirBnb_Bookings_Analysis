```python
# Making a copy of dataframe

new_df = df.copy()
```

```python
sns.boxplot(x = new_df['minimum_nights'])
```

```python
# writing a outlier function for removing outliers in important columns.

def iqr_technique(column):
  Q1 = np.percentile(column,25)
  Q3 = np.percentile(column,75)
  # getting interquartile range
  IQR = Q3 - Q1
  lower_range = Q1 - (1.5 * IQR)
  upper_range = Q3 + (1.5 * IQR)

  return lower_range,upper_range
```

```python
# Removing outliers from column minimum_night

lower_bound,upper_bound = iqr_technique(df['minimum_nights'])

new_df = new_df[(new_df.minimum_nights>lower_bound) & (new_df.minimum_nights<upper_bound)]
```

```python
# Checking Description 

new_df.describe()
```

###**Problem 5.What is the average price preferred by the customer for each neighbourhood_group and for each room type(listing type).**

```python
# Creating a visualization for avg prices of each neighbourhood_group with different room types
pc = sns.barplot(data=avg_pref_price,x='neighbourhood_group',y='price',hue='room_type',palette='bright')
patches = pc.patches
for i in range(len(patches)):
  x = patches[i].get_x() + patches[i].get_width()/2
  y = patches[i].get_height()+.05
  pc.annotate('{:.1f}'.format(avg_pref_price.price[i]),(x,y),ha='center')
plt.show()
```
